```python
# let's investigate dtyi

dtyi = geometry.dty_to_dtyi(dty, ystep, ybincens[0])

fig, ax = plt.subplots()
ax.scatter(omega, dtyi)
ax.set_xlabel("Omega (deg)")
ax.set_ylabel("dtyi")
ax.set_ylim(len(ybincens), 0)

plt.show()
```

```python
# now let's come up with a sinogram image
# we will determine it from ystep and omega

sino = np.zeros((ny, len(omega)), dtype=float)

for i in range(sino.shape[0]):  # ny
    for j in range(sino.shape[1]):  # omega
        # use dtyi_calc as a lookup table from omega index
        this_dtyi = dtyi[j]
        sino[i, j] = 1/(50*np.cbrt((np.abs(i - this_dtyi)))+0.01)

fig, ax = plt.subplots()
ax.imshow(sino, aspect=1)
ax.set_xlabel("Omega step")
ax.set_ylabel("dty step")
plt.show()
```

```python
shift = geometry.sino_shift(y0, ybeam, ystep)

recon = run_iradon(sino, omega, pad=13, shift=shift)

fig, ax = plt.subplots()
ax.imshow(recon, vmin=0, origin="lower")
ax.set_ylabel("i")
ax.set_xlabel("j")
plt.show()
```

```python
step_points = geometry.step_grid_from_ybincens(ybincens, step_size=ystep, gridstep=1, y0=y0)
```

```python
# how big is our step space?
# should be extreme limit in steps
si, sj = step_points[0]
```

```python
# let's pick a point on the edge of the image in the middle
sj = 0
si, sj
```

```python
# compute the dty values for a sinogram
omega = np.arange(0, 361, 1)
dty_calc = geometry.step_omega_to_dty(si, sj, omega, y0=y0, ystep=ystep)
dtyi_calc = geometry.dty_to_dtyi(dty_calc, ystep, ybincens.min())
np.all(([np.any(dtyc < ybincens) for dtyc in dty_calc]))
```

```python
# now let's come up with a sinogram image
# we will determine it from ystep and omega

sino = np.zeros((len(ybincens), len(omega)), dtype=float)

for i in range(sino.shape[0]):  # ny
    for j in range(sino.shape[1]):  # omega
        # use dtyi_calc as a lookup table from omega index
        this_dtyi = dtyi_calc[j]
        sino[i, j] = 1/(50*np.cbrt((np.abs(i - this_dtyi)))+0.01)

fig, ax = plt.subplots()
ax.imshow(sino, aspect=1)
ax.set_xlabel("Omega step")
ax.set_ylabel("dty step")
plt.show()
```

```python
# projection shifts is in units of pixels

shift = geometry.sino_shift(y0, ybeam, ystep)

recon = run_iradon(sino, omega, pad=np.round(shift*2).astype(int), shift=shift)

fig, ax = plt.subplots()
ax.imshow(recon, vmin=0, origin="lower")
ax.set_ylabel("i")
ax.set_xlabel("j")
plt.show()
```
